```r
cassini_df = as.data.frame(Cassini)
cassini_df = subset (cassini_df, select = -classes)
cassini_df = data.matrix(cassini_df, rownames.force = NULL)
#write.csv(cassini_df, "cassini.csv")
```

```r
is.numeric(cassini_df)
```

```r
kmeans_cassini = kmeans(cassini_df, 3)
```

```r
dbscan_cassini = dbscan(cassini_df, 0.5)
```

```r
spectral_cassini = specc(cassini_df, centers=3)
```

```r
# Select set of specific points as initial cluster center
kmeans_cassini = kmeans(cassini_df, centers=cassini_df[c(521,793,836),,drop=FALSE])
plot(Cassini$x, col = cl_class_ids(kmeans_cassini), xlab = "", ylab = "")
```

```r
partition_list = list()
list_i = 1
n_clusters = list(2,3,4,5,6,7,8,9,10)
```

```r
length(n_clusters)
```

```r
initial_centers = read.csv("initial_centers.csv", header=FALSE)
initial_centers = data.matrix(initial_centers, rownames.force = NULL)
```

```r
initial_centers
```

```r
initial_centers[1:n_clusters[[1]],]
```

```r
list_i = 1
while( list_i <= length(n_clusters) ){
    #print(class(n_clusters[[list_i]]))
    #print(n_clusters[[list_i]])
    print(initial_centers[1:n_clusters[[list_i]],])
    list_i = list_i + 1
}
```

```r
partition_list = list()
list_i = 1
while( list_i <= length(n_clusters) ){
    partition_list[[list_i]] <- kmeans(cassini_df, centers=initial_centers[1:n_clusters[[list_i]],], iter.max = 100)
    list_i = list_i + 1
}
```

```r
mcons = cl_consensus(cl_ensemble(list = partition_list), method = 'GV3', weights = 1, control = list())
```

```r
pdf(file="cassini-clue-consensus-scatter-plot.pdf")
plot(Cassini$x, col = cl_class_ids(mcons), xlab = "", ylab = "")
dev.off()
```

```r
list_i = 1
while( list_i <= length(partition_list) ){
    plot(Cassini$x, col = cl_class_ids(partition_list[[list_i]]), xlab = "", ylab = "")
    list_i = list_i + 1
}
```
